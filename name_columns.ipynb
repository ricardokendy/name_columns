```python
def name_columns(df):
  """
  Esta função permite padronizar todos os nomes das colunas de um dataframe, removendo as strings de espaços e pontos. Além de adicionar ou não o _
  df: dataframe que será alterado os nomes das colunas
  return: dataframe corrigido
  Autor: Ricardo Kendy Nakazone
  Versão 1.0
  """
  import pandas as pd
  resp = int(input('Do you prefer 1-(Avg_Area_Income) or 2-(AvgAreaIncome), answer only number: '))
  if resp == 1:
    for k in df.columns:
      word = ""
      word2 = ''
      word3 = ''
      cont = 0
      letter = ''
      for i in k:
        if i == ' ' or i =='.' or i=='-':
          word2 += '_'
        else:
          word2+=i
      word = ''
      for i in word2:
        word+=i
        cont+=1
        if letter =='_' and i == '_':
          for i in range(0,len(word)-1):
            word3 += word[i]
          word = word3
        letter = i
      df = df.rename(columns={k:word})

  elif resp==2:
    for k in df.columns:
      word = ""
      word2 = ''
      word3 = ''
      cont = 0
      letter = ''
      for i in k:
        if i == ' ' or i =='.' or i=='-':
          word2 += '_'
        else:
          word2+=i
      word = ''
      for i in word2:
        word+=i
        cont+=1
        if letter =='_' and i == '_':
          for i in range(0,len(word)-1):
            word3 += word[i]
          word = word3
        letter = i

      word2 = word
      word3 =''
      for i in word2:
        if i !='_':
          word3 += i
      df = df.rename(columns={k:word3})

  return df
```
